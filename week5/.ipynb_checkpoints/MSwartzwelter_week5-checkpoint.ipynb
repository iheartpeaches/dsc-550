```python
#import necessary packages
import pandas as pd
import numpy as np
from textblob import TextBlob 
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from yellowbrick.classifier import ConfusionMatrix
from yellowbrick.classifier import ClassificationReport
from yellowbrick.classifier import ROCAUC
import matplotlib.pyplot as plt


```

```python
#import file note it is a tsv
reviews_train = pd.read_csv('labeledTrainData.tsv', sep='\t')

#convert to lowercase
reviews_train['review'] = reviews_train['review'].apply(lambda x: x.lower())

#create function to remove punctuation
def remove_punctuation(text):
    for punctuation in string.punctuation:
            text = text.replace(punctuation, ' ')
    return text
            
#apply function
reviews_train['review']= reviews_train['review'].apply(lambda x: remove_punctuation(x))

#first tokenize the words in the reviews
reviews_train['tokenized_reviews'] = reviews_train['review'].apply(lambda x: word_tokenize(x))

#create stop words list
stop_words = stopwords.words('english')

#create function to remove stopwords
def remove_stopwords(text):
    text = [word for word in text if word not in stop_words]
    return text

#apply function to tokenized reviews
reviews_train['removed_stopwords'] = reviews_train['tokenized_reviews'].apply(lambda x: remove_stopwords(x))

#create stemmer
porter = PorterStemmer()

#create function to stem words
def stem_words(text):
    text = [porter.stem(word) for word in text]
    listToStr = ' '.join([str(elem) for elem in text])

    return listToStr

#apply function to reviews
reviews_train['apply_stemmer'] = reviews_train['removed_stopwords'].apply(lambda x: stem_words(x))



```

Split this into a training and test set. 


```python
#create input (x) and output (y)
y= reviews_train.sentiment
x=reviews_train.drop('sentiment',axis=1)

#split data
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.2)
```

Fit and apply the tf-idf vectorization to the training set. 


```python
#create the tf-idf matrix
tfidf = TfidfVectorizer()

x_train = tfidf.fit_transform(x_train['apply_stemmer'])
```

Apply but DO NOT FIT the tf-idf vectorization to the test set (Why?). 

```python
#We apply but do not fit the tf-idf vectorization because we do not want to re-train our model on the testing data, we 
#just want to apply the transform
x_test = tfidf.transform(x_test['apply_stemmer']) # apply without fitting



```

Train a logistic regression using the training data. Find the model accuracy on test set. 

```python
#create the logistic regression model 
log_model = LogisticRegression(max_iter = 500)
#fit the model to the training set
log_model.fit(x_train, y_train)

# Making Predictions of y_value
y_test_pred = log_model.predict(x_test)

#Find the accuracy of the model
from sklearn.metrics import accuracy_score
print('Model Accuracy:', accuracy_score(y_test,y_test_pred)*100, '%', sep = '')

```
